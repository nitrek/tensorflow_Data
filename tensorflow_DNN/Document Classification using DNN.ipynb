# Document Classification using DNN

```python
import tensorflow as tf
import pandas as pd
import tempfile
```

```python
train_X = pd.read_csv('/home/praveen/tensorflow_Data/tensorflow/trainX.csv', low_memory=False, header = None, delimiter='\t')
```

```python
train_X.shape[1]
```

```python
train_Y = pd.read_csv('/home/praveen/tensorflow_Data/tensorflow/trainY.csv', low_memory=False, header = None, delimiter='\t')
```

```python
#Converting binary to decimal(0,1,2,3)
train_Y_b10 = pd.DataFrame(train_Y.idxmax(axis=1))
```

```python
train_Y_b10
```

```python
feature_columns = [tf.contrib.layers.real_valued_column("", dimension=train_X.shape[1])]
```

```python
model_dir = tempfile.mkdtemp()
model = tf.contrib.learn.DNNClassifier(feature_columns=feature_columns,
                                            hidden_units=[10,20],
                                            n_classes=4,
                                            model_dir=model_dir)
```

```python
#Fitting the model on the training data
model.fit(train_X,train_Y_b10, steps=1000)
```

```python
results = model.evaluate(train_X,train_Y_b10, steps=1)
print "%s: %s" % ('Accuracy', results['accuracy'])
```

```python
pd.DataFrame(model.predict_proba(train_X))
```

```python
test_X = pd.read_csv('/home/praveen/tensorflow_Data/tensorflow/testX.csv', low_memory=False, header = None, delimiter='\t')
```

```python
test_Y = pd.read_csv('/home/praveen/tensorflow_Data/tensorflow/testY.csv', low_memory=False, header = None, delimiter='\t')
```

```python
test_Y_b10 = pd.DataFrame(test_Y.idxmax(axis=1))
```

```python
results_test = model.evaluate(test_X,test_Y_b10, steps=1)
print "%s: %s" % ('Accuracy', results_test['accuracy'])
```
